## 1.2 Примеры событий

Распределение Пуассона является адекватно моделью, если:
1. События происходят независимо друг от друга;
2. События не могут произойти одовременно;
3. Средняя частота событий постоянна. То есть вероятность попадания того или иного числа событий на любой участок времени длиной t не зависит от начала этого участка на оси времени, а зависит только от его длины t;
4. Количество испытаний достаточно велико, при этом вероятность появления события мала.

### 1.2.1 Интерпретация № 1

Рассмотрим задачу о количестве молекул идеального газа в некотором объеме **V**.

Разобьем $V$ на $n$ объемов $dV$ и примем, что в одном $dV$ может находиться только одна молекула, то есть вероятность нахождения двух и больше молекул в одном объеме $dV$ пренебрежимо мала.

Пусть наблюдения за каждым $dV$ проводятся независимо, то есть факт нахождения в одном $dV$ молекулы никак не влияет на нахождение молекулы в другом $dV$.

Очевидно также, что среднее количество молекул в объеме $V$ пропорционально размеру этого объема. Примем коэффициент пропорциональности за $\lambda$. Следовательно, вероятность нахождения молекулы в конкретном $dV:\;\; p = \dfrac{\lambda}{n}$.

Тогда вероятность нахождения $m$ молекул в объеме $V: \;\; p_{m, n} \sim Bi(n, \dfrac{\lambda}{n})$. Устремим $n$ к бесконечности. Тогда по теореме Пуассона, $p_{m, n} \rightarrow p_m = \dfrac{\lambda^m}{m!}e^{-\lambda}$

### 1.2.2. Интерпретация № 2

Предположим, есть два города, П и В, в каждом есть одна телефонная станция. Необходимо определить, какое минимальное число телефонных линий нужно построить, чтобы только для 1 из 100 вызовов не нашлось бы свободной линии (т.е. вероятность потери звонка не более 0.01).

Пусть каждый день в определенное время некоторые Пятачки из города П созваниваются со своими Виннии-Пухами из города В и договаривается о налете на улей неправильных пчел. Все детали они успевают обсудить в среднем за 2 минуты.

Пускай это "пиковое время" длится T. Разобьем T на маленькие промежутки t, в каждый из которых по правилам городов П и В не может совершаться более одного вызова. Таких промежутков n. Тогда среднее число звонков будет равно $\dfrac{t}{n}\dfrac{\lambda}{2}$.

Получим вероятность, что за время t созвонятся l Пятачков и Винни-Пухов: 
$q_l = C^l_n \left(\dfrac{t}{n}\dfrac{\lambda}{2}\right)^l \left(1-\dfrac{t}{n}\dfrac{\lambda}{2}\right)^{n-l}$

При устремлении числа n к бесконечности будет выполняться теорема Пуассона и $q_l \rightarrow \dfrac{\frac{t\lambda}{2}^l e^ {-\frac{t\lambda}{2}}}{l!}$.

Задача о количестве телефонных линий, поставленная выше, решается с помощью нахождения квантилей распределения.

## 1.3 Моделирование случайной величины $ ^{[1]} $

Пусть $u$ - значение случайной величины $U$, распределенной равномерно на $(0; 1)$, $p_j$ - вероятность значения $a_j$.
Разобьем отрезок $[0; 1]$ на полуинтервалы $\Delta_1 = [0; p_1),\; \Delta_2 = [p_1, p_1 + p_2), \dots ,\Delta_K = [p_1 + \dots + p_{K-1}, 1]$. Необходимо найти интервал $\Delta_j$, в котором лежит $u$, и тогда искомое значение случайной величины, распределенной по дискретному закону, будет $a_j$.

Для распределения Пуассона $p_0 = e^{-\lambda}, \;\; p_j = \dfrac{\lambda^j}{j!}e^{-\lambda}$. Также легко заметить, что $ p_j $ связаны между собой рекуррентным соотношением $p_j = p_{j-1}\dfrac{\lambda}{j}$. 

```python
import random
import math

def rand_Pois(lamb, n):
    nums = []
    for i in range(n):
        p = math.exp(-lamb)
        sp = 0
        sp += p
        u = random.uniform(0,1)
        j = 0
        while u > sp:
            j += 1         # p_j
            p *= lamb / j  # p_j / p_j-1 = lamb / j
            sp += p
        nums.append(j)
    return nums

print('Для λ = 1:', rand_Pois(1, 150))
```

## 2. Непрерывное распределение Рэлея

| Закон распределения (плотность вероятностей): | $ f(x, \sigma) = \dfrac{x}{\sigma^2}exp\left(-\dfrac{x^2}{2\sigma^2}\right), \;\;x >= 0, \sigma > 0 $ |
| ----------------------------:|:-------------------------------------------:|

## 2.1 Описание основных математических характеристик
|                              |                                             |
| ----------------------------:|:-------------------------------------------|
| Функция распределения:       |$ 1-exp\left(-\dfrac{x^2}{2\sigma^2}\right) $|
| Математическое ожидание      | $ \sqrt{\dfrac{\pi}{2}}\sigma $             |
| Дисперсия:                   | $ \left(2-\dfrac{\pi}{2}\right)\sigma^2 $   | 
| Характеристическая функция:  | $ 1 + t\sigma\sqrt{\dfrac{\pi}{2}}e^{\frac{\sigma^2 t^2}{2}}\cdot\left[\dfrac{2}{\sqrt{\pi}}\int\limits_0^{\frac{t \sigma}{\sqrt{2}}} e^{-t^2}dt +1  \right] , \;\; \dfrac{2}{\sqrt{\pi}}\int\limits_0^{\frac{t \sigma}{\sqrt{2}}} e^{-t^2}dt = erf\left(\dfrac{t \sigma}{\sqrt{2}}\right)$    |
| Производящая функция:        | $ 1 - t\sigma\sqrt{\dfrac{\pi}{2}}e^{-\frac{\sigma^2 t^2}{2}}\cdot\left[erfi\left(\dfrac{t \sigma}{\sqrt{2}}\right) -i  \right] $     |

#### Выводы:

- $E\xi = \int\limits_0^\infty x \dfrac{x}{\sigma ^ 2} e^{-\frac{x^2}{2\sigma^2}} dx = 
 \dfrac{1}{\sigma^2}\int\limits_0^\infty x \cdot x \cdot e^{-\frac{x^2}{2\sigma^2}} dx  =
 \left\{u = x, \;du = dx, \; v = \int x e^{-\frac{x^2}{2\sigma^2}},\; dv = x e^{-\frac{x^2}{2\sigma^2}} dx\right\} = \\ =
\dfrac{1}{\sigma^2}\left[ x\cdot \int x e^{-\frac{x^2}{2\sigma^2}} dx \big|_0^\infty -  
\int_0^\infty x e^{-\frac{x^2}{2\sigma^2}} dx \right]  
= \left\{ \int x e^{-bx^2} dx = \dfrac{1}{2b} e^{-bx^2} \right\} = 
\dfrac{1}{\sigma^2} \left[ x \cdot \dfrac{1}{2} (-2\sigma^2) e^{-\frac{x^2}{2\sigma^2}} \bigg|_0^\infty - 
\int_0^\infty \dfrac{1}{2}(-2\sigma^2)e^{-\frac{x^2}{2\sigma^2}}dx\right] = \\ =
0 + \int\limits_0^\infty e^{-\frac{x^2}{2\sigma^2}}dx = \dfrac{\sqrt{2 \pi \sigma^2}}{\sqrt{2 \pi \sigma^2}}\cdot \int\limits_0^\infty e^{-\frac{x^2}{2\sigma^2}} dx = \dfrac{\sqrt{2 \pi \sigma^2}}{2} = \sigma\sqrt{\dfrac{\pi}{2}}$


- $E(\xi)^2 = \int\limits_0^\infty x^2 \dfrac{x}{\sigma ^ 2} e^{-\frac{x^2}{2\sigma^2}} dx =
\left\{u = x^2, \; du = 2dx, \; v = \int x e^{-\frac{x^2}{2\sigma^2}}, \; dv = x e^{-\frac{x^2}{2\sigma^2}} dx\right\} =
x^2\dfrac{1}{2\sigma^2}\left(-2\sigma^2\right) e^{-\frac{x^2}{2\sigma^2}} \bigg|_0^\infty - \\ -
\int\limits_0^\infty 2x \dfrac{1}{2\sigma^2}\dfrac{1}{2}\left(-2\sigma^2\right) e^{-\frac{x^2}{2\sigma^2}} dx =
0 + 2 \int\limits_0^\infty x e^{-\frac{x^2}{2\sigma^2}} dx =
2 \left(-2\sigma^2\right) \dfrac{1}{2} e^{-\frac{x^2}{2\sigma^2}} \bigg|_0^\infty = 
0 - \left(-2\sigma^2\right) = 2\sigma^2$

- $D\xi = E(\xi)^2-(E\xi)^2 = 2\sigma^2 - \left(\sigma\sqrt{\dfrac{\pi}{2}}\right)^2 = 2\sigma^2 - \sigma^2\dfrac{\pi}{2} =  \left(2-\dfrac{\pi}{2}\right)\sigma^2$

#### График плотности:

```python
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline

def f_rayl(x, sigma):
    return math.exp(-(x ** 2) / (2 * sigma ** 2)) * (x / (2 * sigma ** 2))

my_simple_plot(range(30), [1,2,3,5,8], f_rayl, False)
plt.xlabel = 'x'
plt.ylabel = 'f(x)'
plt.gca().legend(('σ = 1','σ = 2', 'σ = 3', 'σ = 5', 'σ = 8'))
plt.title('График плотности распределения')
```

## 2.2 Примеры событий

### 2.2.1 Типичная интерпретация

Типичной ситуацией, где возникает распределение Рэлея, является стрельба из лука.

Пусть центр мишени -- это начало координат. Пусть вектор, соединяющий начало координат и точку попадания стрелы, имеет координаты $(u, v)$, которые распределены по нормальному закону, следовательно, $ f_u(x) = f_v(x) = \dfrac{1}{\sigma \sqrt{2\pi}} exp\left(-\dfrac{x^2}{2\sigma^2}\right) $. Точность попадания в мишень определяет длина вектора: чем он короче, тем точнее был выстрел. Пусть $ L = \sqrt{u^2+v^2} $ -- длина вектора. Пусть круг $ D_x = \{(u,v): L < x\} $ -- область, которая получится при повороте вектора. Тогда функция распределения для двух величин $u, v: \;\; F_x(x) = \iint_{D_x} f_u(x)f_v(x)dx = \dfrac{1}{2\pi\sigma^2}\int\limits_0^{2\pi}\int\limits_0^x r e^{\frac{-r^2}{2\sigma^2}} dr d\phi = 
\dfrac{1}{\sigma^2} \int\limits r e^{\frac{-r^2}{2\sigma^2}} dr$. По определению, функция плотности $ f(x, \sigma) = F'(x, \sigma) = \dfrac{x}{\sigma^2} e^{-\frac{x^2}{2\sigma^2}}$. Получена функция плотности вероятности для распределения Рэлея.

Приведем более конкретный пример (на основе примера из источника [3]).  Пусть лучник стреляет, имея стандартную мишень диаметром 40 см. Пусть среднеквадратичное отклонение от центра мишени одинаково для обеих координат точки попадания стрелы: $ \sigma_u = \sigma_v=8см. $ Тогда функция плотности вероятности $ f(x) = \dfrac{x}{64} e^{-\frac{x^2}{2 \cdot 64}}$. Математическое ожидание: $ E\xi =  \sqrt{\dfrac{\pi}{2}}\cdot 8 = 10.0265$. Найдем вероятность попадания в "десятку", диаметр которой равен 8 см: $ F(4) = 0.117503  $. ## 2.3 Моделирование случайной величины

Моделирование случайных величины, распределенных по произвольному закону (в данному случае по закону Рэлея) может быть основано на генерации равномерно распределенных случайных величин.

Используем аналитический метод, то есть метод обратной функции. Построим зависимость значений случайной величины, распределенной по закону Рэлея, от равномерно распределенной величины, лежащей в интервале от 0 до 1:

$ P(\xi < t) = t\;\; \forall t \in (0, 1) $ -- равномерное распределение

$ P(\xi < F(x)) = P(F^{-1}(\xi) < x) = F_{F^{-1}(\xi)}(x) = F(x)$

$ F(F^{-1}(x)) = x $

$ F^{-1} = \sigma \sqrt{-2 ln U} $, где $U$ - равномерно распределенная случайная величина $^{[3]}$

```python
import random
import math

def inv_f_Rayl(sigma, u):
    return sigma * math.sqrt(-2 * math.log(u))

def rand_Rayl(sigma, n):
    nums = []
    for i in range(n):
        u = random.uniform(0, 1)
        nums.append(inv_f_Rayl(sigma, u))
    return nums

nums_of_Rayl = rand_Rayl(3, 15)
print('Для 𝜎 = 3:')
print(*nums_of_Rayl, sep = '; ')
```
